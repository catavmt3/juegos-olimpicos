**Tipos de datos**

Incluímos aquí las características originales del conjunto de datos, con su descripción y tipo.

  **ID** - Número único para cada atleta, int64
  
  **Name** (**Nombre**) - Nombre del atleta, object
  
  **Sex** (**Genero**) - M (Masculino) o F (Femenino), object
  
  **Age** (**Edad**) - Entero que representa la edad del atleta, float64
  
  **Height** (**Altura**) - Altura del atleta en centímetros, float64
  
  **Weight** (**Peso**) - Peso del atleta en kilogramos, float64
  
  **Team** (**Equipo**) - Nombre del equipo al que pertenece el atleta, object
  
  **NOC** - Código de 3 letras del Comité Olímpico Nacional del altleta, object
  
  **Games** (**Juegos**) - Año y temporada de los Juegos Olímpicos, object
  
  **Year** (**Año**) - Año de los Juegos Olímpicos, int64
  
  **Season** (**Temporada**) - Verano o Invierno, object
  
  **City** (**Ciudad**) - Ciudad anfitriona de los Juegos Olímpicos, object
  
  **Sport** (**Deporte**) - Deporte en el que compitió el atleta, object
  
  **Event** (**Evento**) - Evento específico en el que compitió el atleta, object
  
  **Medal** (**Medalla**) - Oro, Plata, Bronce o NA (no ganó medalla), object

```python
# Intervalos entre las distintas ediciones de verano
years_diff_summer = sorted(JJOO[JJOO['Temporada'] == 'Summer']['Año'].dt.year.unique())

# Mostrar los años que no cumplen el intervalo de 4 años
for i in range(1, len(years_diff_summer)):
    if years_diff_summer[i] - years_diff_summer[i-1] != 4:
        print(f'Diferencia de {years_diff_summer[i] - years_diff_summer[i-1]} años entre {years_diff_summer[i-1]} y {years_diff_summer[i]}')
```

***Explicación de las diferencias***

1904-1906: La edición de 1906 fue un evento "intercalado" que fue más bien una edición no oficial (Juegos Olímpicos Intercalados) y no parte del ciclo normal de 4 años.

1906-1908: La diferencia de 2 años se debió a ajustes logísticos y la organización de los Juegos Olímpicos de Londres de 1908.

1912-1920: La Primera Guerra Mundial causó la cancelación de los Juegos Olímpicos de 1916, lo que resultó en una diferencia de 8 años entre 1912 y 1920.

1936-1948: La Segunda Guerra Mundial causó la cancelación de los Juegos Olímpicos de 1940 y 1944, lo que resultó en una diferencia de 12 años entre los Juegos Olímpicos de 1936 y 1948.

```python
# Intentaremos clasificar las disciplinas de acuerdo a si es un deporte individual o en equipo mediante el uso de palabras clave.

# Definimos diccionario con palabras clave que identifican deportes en equipo
team_keywords = {'team', 'relay', 'doubles', 'duet', 'group', 'pairs', 'fours', 'synchronized','two person'}
# Definimos diccionario con deportes en equipo
pure_team_sports = {
    'baseball',
    'basketball',
    'cricket',
    'hockey',
    'polo',
    'rugby sevens',
    'rugby',
    'softball',
    'volleyball',
    'beach volleyball',
    'football',
    'soccer',
    'handball',
    'bobsleigh'
}

def classify_event(row):
    event = row['event'].lower()
    # si en el evento se encuantra alguna de las palabras clave o se trata de uno de los previamente asignamos el valor Team, de lo contrario individual
    if any(kw in event for kw in team_keywords.union(pure_team_sports)):
        return 'Team'
    return 'Individual'

# aplicamos para agregar la columna type
JJOO['type'] = JJOO.apply(classify_event, axis=1)

# preview results
print(JJOO[['event', 'type']].drop_duplicates().head(10))
```

```python
# Calculamos las Medidas de Posición

# Función para calcular medidas de posición
def calcular_medidas_posicion(datos):
    """Calcula y muestra las principales medidas de posición."""
    resultados = pd.DataFrame({
        'Media': datos.mean(),
        'Mediana': datos.median(),
        'Moda': datos.mode().iloc[0],  # Toma el primer valor si hay múltiples modas
        'Q1 (25%)': datos.quantile(0.25),
        'Q3 (75%)': datos.quantile(0.75),
        'Mínimo': datos.min(),
        'Máximo': datos.max()
    })

    return resultados

# Aplicamos la función a nuestro dataset
medidas_posicion = calcular_medidas_posicion(JJOO_num)
print("Medidas de Posición:")
print(medidas_posicion)
```

La mayoría de los atletas olímpicos se encuentran en un rango de edad relativamente joven, entre los 22 y 28 años.

La altura de los atletas olímpicos se distribuye en torno a un promedio de 175 cm, con una mayor frecuencia de atletas que miden 180 cm.

El peso de los atletas olímpicos también se distribuye en torno a un promedio de 70 kg, con una mayor frecuencia de atletas que pesan 70 kg. La variabilidad en el peso puede estar relacionada con las diferentes exigencias físicas de cada deporte.

Las medidas de posición son herramientas útiles para describir la distribución de los datos, pero es importante complementarlas con otras medidas, como las de dispersión, para tener una visión más completa.


```python
# Calculamos Medidas de Dispersión

# Función para calcular medidas de dispersión
def calcular_medidas_dispersion(datos):
    """Calcula y muestra las principales medidas de dispersión."""
    resultados = pd.DataFrame({
        'Rango': datos.max() - datos.min(),
        'Varianza': datos.var(),
        'Desviación Estándar': datos.std(),
        'Coef. de Variación (%)': (datos.std() / datos.mean()) * 100,
        'IQR': datos.quantile(0.75) - datos.quantile(0.25),
        #'MAD': datos.mad()  # Desviación absoluta media
    })

    return resultados

# Aplicamos la función a nuestro dataset
medidas_dispersion = calcular_medidas_dispersion(JJOO_num)
print("\nMedidas de Dispersión:")
print(medidas_dispersion)
```

La edad de los atletas olímpicos presenta una gran variabilidad, con un rango amplio y una dispersión considerable alrededor de la media.

La altura de los atletas olímpicos presenta una menor variabilidad en comparación con la edad, con un rango y una dispersión más reducidos.

El peso de los atletas olímpicos presenta una gran variabilidad, similar a la de la edad, con un rango amplio y una dispersión considerable alrededor de la media.

La variabilidad en la edad, altura y peso puede estar influenciada por diversos factores, como la disciplina deportiva, el género y la nacionalidad de los atletas.

El coeficiente de variación nos permite comparar la variabilidad de conjuntos de datos con diferentes unidades o escalas, como en nuestro caso.

```python
# Calculamos Medidas de Forma

# Función para calcular medidas de forma
def calcular_medidas_forma(datos):
    """Calcula y muestra las principales medidas de forma."""
    resultados = pd.DataFrame({
        'Asimetría': datos.skew(),
        'Curtosis': datos.kurtosis()
    })

    return resultados

# Aplicamos la función a nuestro dataset
medidas_forma = calcular_medidas_forma(JJOO_num)
print("\nMedidas de Forma:")
print(medidas_forma)
```

La edad y el peso presentan distribuciones sesgadas a la derecha, con una mayor concentración de atletas en los valores más bajos y una menor cantidad en los valores más altos. Además, ambas variables tienen una curtosis positiva (leptocúrtica), lo que indica una mayor concentración de datos alrededor de la media y una mayor cantidad de valores extremos.

```python
# Detección de outliers

Definimos una función para detectar outliers en un conjunto de datos.
def detectar_outliers(datos, columna, metodo='iqr'):
    """
    Detecta valores atípicos utilizando el método IQR o Z-score.

    Args:
        datos: DataFrame de pandas
        columna: Nombre de la columna a analizar
        metodo: 'iqr' o 'zscore'

    Returns:
        DataFrame con los outliers identificados
    """
    if metodo == 'iqr':
        Q1 = datos[columna].quantile(0.25)
        Q3 = datos[columna].quantile(0.75)
        IQR = Q3 - Q1

        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR

        outliers = datos[(datos[columna] < limite_inferior) |
                          (datos[columna] > limite_superior)]

        print(f"Método IQR para {columna}:")
        print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
        print(f"Límite inferior: {limite_inferior:.2f}")
        print(f"Límite superior: {limite_superior:.2f}")

    elif metodo == 'zscore':
        z_scores = stats.zscore(datos[columna])
        outliers = datos[abs(z_scores) > 3]

        print(f"Método Z-score para {columna}:")
        print("Valores con |z| > 3 son considerados outliers")

    print(f"Número de outliers encontrados: {len(outliers)}")
    return outliers

# Aplicamos la función a nuestros datos
for columna in JJOO_num.columns:
    print(f"\nDetección de outliers en {columna}:")
    outliers_iqr = detectar_outliers(JJOO_num, columna, 'iqr')
    outliers_zscore = detectar_outliers(JJOO_num, columna, 'zscore')
```

Se observa la presencia de outliers en las tres variables, aunque en diferentes cantidades según el método de detección utilizado.

Los outliers en la edad podrían deberse a la participación de atletas muy jóvenes o muy veteranos en ciertas disciplinas. Los outliers en la altura y el peso podrían estar relacionados con las diferentes exigencias físicas de cada deporte, así como con las características individuales de los atletas.

```python
# Shapiro-Wilk, que no aplica por la cantidad de datos

def visualizar_distribucion(datos, columna):
    """Visualiza la distribución de una variable con histograma, boxplot y QQ plot."""
    fig, axes = plt.subplots(1, 3, figsize=(20, 6))

    # Histograma con curva de densidad
    sns.histplot(datos[columna], kde=True, ax=axes[0])
    axes[0].axvline(datos[columna].mean(), color='red', linestyle='--', label='Media')
    axes[0].axvline(datos[columna].median(), color='green', linestyle='-.', label='Mediana')
    axes[0].set_title(f'Histograma de {columna}')
    axes[0].legend()

    # Boxplot
    sns.boxplot(x=datos[columna], ax=axes[1])
    axes[1].set_title(f'Boxplot de {columna}')

    # QQ plot
    stats.probplot(datos[columna], plot=axes[2])
    axes[2].set_title(f'Q-Q Plot de {columna}')

    plt.tight_layout()
    plt.show()

    # Mostrar estadísticas resumidas
    print(f"\nEstadísticas resumidas para {columna}:")
    print(datos[columna].describe())

    # Mostrar medidas de forma
    print(f"\nAsimetría: {datos[columna].skew():.4f}")
    print(f"Curtosis: {datos[columna].kurtosis():.4f}")

    # Prueba de normalidad
    stat, p_valor = stats.shapiro(datos[columna])
    print(f"\nPrueba de normalidad Shapiro-Wilk:")
    print(f"Estadístico: {stat:.4f}, p-valor: {p_valor:.4e}")
    if p_valor < 0.05:
        print("Los datos no siguen una distribución normal (rechaza H0)")
    else:
        print("No hay evidencia para rechazar la normalidad de los datos")

# Ejemplo de uso para cada columna
for columna in JJOO_num.columns:
    visualizar_distribucion(JJOO_num, columna)
```

```python
# Función para aplicar múltiples tests de normalidad
def test_normalidad(datos, nombre=None):
    """
    Aplica múltiples tests de normalidad a un conjunto de datos
    y devuelve los resultados en un DataFrame.
    """
    # Anderson-Darling test
    anderson_test = stats.anderson(datos, dist='norm')

    # Kolmogorov-Smirnov test
    ks_test = stats.kstest(datos, 'norm', args=(datos.mean(), datos.std()))

    # Crear DataFrame con resultados
    resultados = pd.DataFrame({
        'Test': ['Kolmogorov-Smirnov'],
        'Estadístico': [ks_test[0]],
        'p-valor': [ks_test[1]],
        'Normalidad (α=0.05)': [
            'Aceptada' if ks_test[1] > 0.05 else 'Rechazada'
        ]
    })

    # Añadir resultados de Anderson-Darling
    # (este test tiene múltiples valores críticos)
    ad_result = 'Rechazada'
    for i, nivel in enumerate([15, 10, 5, 2.5, 1]):
        if anderson_test[0] < anderson_test[1][i]:
            ad_result = f'Aceptada (α={nivel}%)'
            break

    ad_row = pd.DataFrame({
        'Test': ['Anderson-Darling'],
        'Estadístico': [anderson_test[0]],
        'p-valor': [None],  # Anderson-Darling no reporta p-valor directamente
        'Normalidad (α=0.05)': [ad_result]
    })

    resultados = pd.concat([resultados, ad_row], ignore_index=True)

    if nombre:
        print(f"Resultados para: {nombre}")

    return resultados

# Aplicar los tests a cada conjunto de datos
resultados = {}
for columna in JJOO_num.columns:
    resultados[columna] = test_normalidad(JJOO_num[columna], columna)
    print(resultados[columna])
    print("\n" + "="*80 + "\n")
```

Si bien notamos que la altura es la que más se acerca a una distribución normal, el test rechaza la hipótesis de normalidad.

```python
import plotly.express as px
import pandas as pd
import numpy as np
import os
import shutil
import imageio.v2 as imageio
from PIL import Image, ImageDraw, ImageFont

# --- 1. Preparar datos ---
filtered_JJOO = JJOO[(JJOO['season'] == 'Summer') & (JJOO['year'] >= 1984)]
grouped = filtered_JJOO.groupby(['year', 'sport'])['id'].count().reset_index()
grouped = grouped.rename(columns={'id': 'count'})
grouped['year'] = grouped['year'].astype(float)

# --- 2. Interpolación suave ---
years = sorted(grouped['year'].unique())
all_sports = grouped['sport'].unique()

interp_years = []
for i in range(len(years) - 1):
    interp_years += list(np.linspace(years[i], years[i + 1], 5, endpoint=False))
interp_years.append(years[-1])
interp_years = [round(y, 2) for y in interp_years]

full_data = pd.DataFrame([
    {'year': y, 'sport': s} for y in interp_years for s in all_sports
])
merged = full_data.merge(grouped, how='left', on=['year', 'sport'])
merged = merged.sort_values(by=['sport', 'year'])
merged['count'] = merged.groupby('sport')['count'].transform(lambda g: g.interpolate().fillna(0))

# --- 3. Crear carpeta de frames ---
os.makedirs("frames", exist_ok=True)
filenames = []

# --- 4. Generar frames + overlay con PIL ---
font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf"  # Cambia si usas otro sistema
year_font_size = 60  # Tamaño para el año
axis_font_size = 14 # Tamaño para los ejes <-- Nuevo parámetro
progress_height = 10

# Colores personalizados por deporte
custom_colors = {
    'Athletics': '#4D4D4D',     # Gris oscuro
    'Gymnastics': '#F39200',    # Naranja suave
    'Swimming': '#76B7B2',      # Verde-azulado
    'Rowing': '#E15759',        # Rojo suave
    'Shooting': '#B07AA1',      # Violeta
    'Weightlifting': '#FF9DA7', # Rosa viejo
    'Fencing': '#59A14F',       # Verde
    'Judo': '#EDC948',          # Amarillo mostaza
    'Diving': '#BAB0AC',        # Gris claro
    'Wrestling': '#9C755F'      # Marrón claro
}

for i, year in enumerate(interp_years):
    frame = merged[merged['year'] == year]

    fig = px.bar(frame, x='sport', y='count', color='sport',
                 title=f'Participación por Deporte - {int(year)}',
                 color_discrete_map=custom_colors)

    # Ajustar el tamaño de la fuente de los ejes
    fig.update_layout(
        xaxis=dict(tickfont=dict(size=axis_font_size)),
        yaxis=dict(tickfont=dict(size=axis_font_size),
                   title=dict(font=dict(size=axis_font_size)))
    )


    # Guardar el frame de Plotly como imagen
    frame_path = f"frames/frame_{i:04d}.png"
    fig.write_image(frame_path, width=1000, height=600)

    # Abrir la imagen del frame con PIL
    img = Image.open(frame_path)
    draw = ImageDraw.Draw(img)
    try:
        # Cargar la fuente para el año
        font = ImageFont.truetype(font_path, year_font_size)
    except IOError:
        # Si la fuente predeterminada no se encuentra, usar una fuente genérica
        print(f"Advertencia: La fuente '{font_path}' no se encontró. Usando fuente genérica.")
        font = ImageFont.load_default()
        year_font_size = 16 # Ajustar tamaño si se usa fuente genérica

    # Posición del año en el frame
    text = str(int(year))
    text_bbox = draw.textbbox((0, 0), text, font=font)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]

    # Coloca el año en la esquina superior derecha (ajusta según necesites)
    x_pos = img.width - text_width - 20
    y_pos = 20

    # Dibuja el año en la imagen
    draw.text((x_pos, y_pos), text, font=font, fill=(0, 0, 0))  # Color negro para el texto

    # Dibuja la barra de progreso
    progress_width = int((i / len(interp_years)) * img.width)
    draw.rectangle([(0, img.height - progress_height), (progress_width, img.height)], fill=(0, 128, 0)) # Barra verde

    # Guarda la imagen con el overlay
    img.save(frame_path)
    filenames.append(frame_path)


# --- 5. Crear GIF a partir de los frames ---
with imageio.get_writer('olympic_participation_animation.gif', mode='I', duration=100) as writer: # duration en ms
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)

# --- 6. Limpiar frames ---
shutil.rmtree("frames")
```
